```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")
```

```python
# Figure 3A
df = pd.read_csv(r'/media/shuang/data/repository/result/DGE_6disorders/disorder1.csv')
#df = pd.read_csv(r'/media/shuang/data/repository/result/DGE_6disorders/disorder2.csv')
f, ax = plt.subplots(figsize=(6, 6), sharex=True, sharey=True)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
sns.kdeplot(data=df, x='x', y='y', hue='hue',palette={1:'#c51b7d', 2:'#4d9221'},fill=True, alpha=0.9,levels=50)
sns.scatterplot(data=df, x='x', y='y', s=20, color=".15")
sns.regplot(data=df, x='x', y='y',scatter=False,truncate=False,color='black')
ax.set(xlim=(-1, 1), ylim=(-1, 1))
plt.savefig("/media/shuang/data/repository/figures/Disorder1.svg", format="svg",dpi=1000)
```

```python
# Figure 3B
data = pd.read_csv(r'/media/shuang/data/repository/result/HARgene_corr.csv')
f, ax = plt.subplots(figsize=(6, 6), sharex=True, sharey=True)
sns.scatterplot(data=data, x='FCP', y='epifc', s=25, color=".15")
sns.regplot(data=data, x='FCP', y='epifc',scatter=False,truncate=False,color='#b2182b')
ax.set(xlim=(-2.5, 2.5), ylim=(-0.3, 0.3))
plt.savefig("/media/shuang/data/repository/figures/FCP.svg", format="svg",dpi=1000)
```
