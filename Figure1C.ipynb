```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
sns.set_theme(style="whitegrid")
```

```python
df = pd.read_csv(r'/media/shuang/data/repository/result/epi400.csv')
df
```

```python
df.yeo17=df.yeo17.astype(str)
hue_order=df.groupby('yeo17')['fcrank'].mean().sort_values().index
keys=hue_order
values=sns.color_palette('coolwarm',17)
sns.catplot(data=df, x="fcrank", y="yeo17",hue="yeo17",order=hue_order, jitter=False, alpha=0.5, s=80, height=8, aspect=0.4, palette=dict(zip(keys,values)))
plt.savefig("/media/shuang/data/repository/figures/fcN17.svg", format="svg",dpi=1000)
```

```python
sns.set_theme(style="whitegrid")
df.yeo17=df.yeo17.astype(str)
hue_order=df.groupby('yeo17')['scrank'].mean().sort_values().index
keys=hue_order
values=sns.color_palette('coolwarm',17)
sns.catplot(data=df, x="scrank", y="yeo17",hue="yeo17",order=hue_order, jitter=False, alpha=0.5, s=80, height=8, aspect=0.4, palette=dict(zip(keys,values)))
plt.savefig("/media/shuang/data/repository/figures/scN17.svg", format="svg",dpi=1000)
```
